```python
# Portanto, correlacionar por nome do município é impossível. Precisamos correlacionar pelo código do IBGE e pelo código do TSE
# Vemos que cada município tem um código nos dois dataframes
# No GeoDataFrame, o código é um código do IBGE
# No DataFrame eleitoral, o código é o código do TSE
# Vamos procurar o código no DataFrame de dados de votação

df_mais_votado_por_municipio[df_mais_votado_por_municipio['NM_MUNICIPIO'] == 'CAIUÁ']
```

```python
# Vamos comparar com o código do IBGE:
mapa_br[mapa_br['NM_MUNICIP'] == 'CAIUÁ']['CD_GEOCMU'].values[0]  == df_mais_votado_por_municipio[df_mais_votado_por_municipio['NM_MUNICIPIO'] == 'CAIUÁ']['CD_MUNICIPIO'].values[0]
```

```python
# Portanto, o link que posso usar entre os dois conjuntos de dados é a coluna de códigos
# Entretanto, o IBGE usa um código e o TSE outro código, preciso então de uma equivalência entre eles
# Existe no GitHub do Estadão um repositório que apresenta essa correlação entre código IBGE e código TSE
# Vamos recuperar essa equivalência direto do Github do estadão

link_equivalencia_tse_ibge = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'
df_equi_tse_ibge = pd.read_csv(link_equivalencia_tse_ibge)
df_equi_tse_ibge.to_excel('equi_tse_ibge.xlsx')
```

```python
# Mudando os nomes das colunas nos dois dataframes para ficarem iguais ao dataframe de equivalência
# Isso facilita a junção dos dataframes

df_mais_votado_por_municipio.rename(columns={'CD_MUNICIPIO': 'COD_TSE'}, inplace=True)
mapa_br.rename(columns={'CD_GEOCMU': 'GEOCOD_IBGE'}, inplace=True)

# Juntando o dataframe de equivalência ao dataframe de votação
df_final = pd.merge(df_equi_tse_ibge, df_mais_votado_por_municipio, on='COD_TSE')
df_final.head()
```

```python
# Tenho colunas repetidas e desnecessárias, vou dropá-las

df_final.drop(columns=['chave', 'NOME', 'UF', 'AJUSTE'], inplace=True)
df_final.head()
```

```python
# Vamos fazer o trabalho para os outros candidatos
# Criando uma função para fazer os mapas para os outros candidatos:

def cria_mapa_vot_eleitor_por_candidato(candidato,
                                        df_turno=df_1o_turno, df_eleitorado=df_eleitorado_f, 
                                        df_eq=df_equi_tse_ibge, 
                                        df_mapa=mapa_br):
    df_cand = df_1o_turno[df_1o_turno['NM_CANDIDATO'] == f'CANDIDATO {candidato}']
    df_cand = pd.merge(df_cand, df_eleitorado_f, on='CD_MUNICIPIO')
    df_cand.rename(columns={'SG_UF_x': 'SG_UF', 
                          'NM_MUNICIPIO_x': 'NM_MUNICIPIO', 
                          'QT_ELEITORES_PERFIL': 'QT_ELEITORES',
                          'CD_MUNICIPIO': 'COD_TSE'}, inplace=True)
    df_cand.drop(columns=['SG_UF_y', 'NM_MUNICIPIO_y'], inplace=True)
    df_cand['VOTOS/ELEITORES'] = df_cand['QT_VOTOS_NOMINAIS']/df_cand['QT_ELEITORES']
    df_cand['VOTOS/ELEITORES (%)'] = (df_cand['QT_VOTOS_NOMINAIS']/df_cand['QT_ELEITORES']) * 100
    df_cand= pd.merge(df_cand, df_equi_tse_ibge, on='COD_TSE')
    df_cand.drop(columns=['NOME', 'UF'], inplace=True)
    df_cand = pd.merge(df_cand, mapa_br, on='GEOCOD_IBGE')
    df_cand.drop(columns='NM_MUNICIP', inplace=True)
    df_cand = gpd.GeoDataFrame(df_cand)
    fig, ax = plt.subplots(figsize=(12, 12))
    df_cand.plot(column='VOTOS/ELEITORES (%)', cmap='Blues', ax=ax)
    plt.title(f'Relação votos/eleitores (%) - Candidato {candidato} - 1º Turno - Eleições 2018', fontdict={'fontsize': '20', 'fontweight': '3'})
    ax.axis('off')
    ax.annotate('Fonte: IBGE, 2015; TSE, 2019; Jornal o Estado de São Paulo, 2019', 
            xy=(0.08, 0.14), xycoords='figure fraction', horizontalalignment='left',
            verticalalignment='top', fontsize=10, color='#555555')

    vmin_rel_vot_elei, vmax_rel_vot_elei = df_cand['VOTOS/ELEITORES (%)'].min(), df_cand['VOTOS/ELEITORES (%)'].max()
    sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin_rel_vot_elei, vmax=vmax_rel_vot_elei))
    cbar = fig.colorbar(sm, shrink=0.5)

    fig.savefig(f'votação {candidato}', dpi=150)
    plt.show()
```
